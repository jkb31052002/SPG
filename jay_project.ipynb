```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Copy of Survey On Social Media Usage (Responses) - Form Responses 1.csv")
```

```python
df.info()
```

```python


import matplotlib.pyplot as plt

# Plot the distribution of a particular column
df['column_name'].value_counts().plot(kind='bar')
plt.show()

# Plot the relationship between two columns
plt.scatter(df['column_1'], df['column_2'])
plt.xlabel('column_1')
plt.ylabel('column_2')
plt.show()

# Create a pie chart of a particular column
df['column_name'].value_counts().plot(kind='pie')
plt.show()
```

```python
df['Gender'].value_counts().plot(kind='pie')
plt.show()
```

```python


df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

```python
df['Age'].value_counts().plot(kind='bar', color='orange')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')

# Display the percentage values on top of each bar
for index, value in enumerate(df['Age'].value_counts()):
    plt.text(index, value + 1, f'{value / len(df["Age"]) * 100:.1f}%', ha='center', va='bottom')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'

plt.figure(figsize=(10, 6))
sns.stripplot(x='Gender', y='Whatâ€™s your Education Level?', data=df, jitter=True, palette='viridis')
plt.title('Scatter Plot of Gender and Education Level')
plt.xlabel('Gender')
plt.ylabel('Whatâ€™s your Education Level?')
plt.show()
```

```python
df['Gender'].value_counts().plot(kind='bar', color=['skyblue', 'pink'])
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(x='How often do you use social media platforms in a typical day?', y='Age', data=df, palette='pastel')
plt.title('Age vs. Social Media Usage Frequency')
plt.xlabel('Social Media Usage Frequency')
plt.ylabel('Age')
plt.show()
```

```python
gender_social_media = df.groupby(['Gender', 'How often do you use social media platforms in a typical day?']).size().unstack()
gender_social_media.plot(kind='bar', stacked=True, color=['lightblue', 'lightcoral', 'lightgreen'])
plt.title('Gender Distribution in Social Media Usage Frequency')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Create a new DataFrame to store the counts for each combination
edu_level_vs_usage = df.groupby(['Whatâ€™s your Education Level?', 'How often do you use social media platforms in a typical day?']).size().unstack()

# Plotting a grouped bar chart
ax = edu_level_vs_usage.plot(kind='bar', figsize=(12, 8), colormap='viridis')

# Adding data labels
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height}', (x + width/2, y + height), ha='center', va='bottom', fontsize=8, color='black')

plt.title('Education Level vs. Social Media Usage Frequency')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='Gender', y='Age', data=df, palette='pastel')

# Adding data labels next to data points
medians = df.groupby('Gender')['Age'].median().values
nobs = df['Gender'].value_counts().values
nobs = [f"n: {i}" for i in nobs]

for i in range(len(medians)):
    plt.text(i, medians[i] + 0.1, nobs[i], ha='center', va='bottom', color='black', fontsize=8)

plt.title('Gender vs. Age')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()

```
